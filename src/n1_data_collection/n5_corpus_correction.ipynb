# Importación de librerías

```python
from pathlib import Path
import os
import re
```

# Importación y transformación de datos

Cargamos las rutas, tanto de entrada (en la carpeta `data/raw/corpus`) como de salida (en la carpeta `data/processed/corpus`).

```python
DATA_ROOT = Path("E:/proyecto/data")
INPUT_DIR = DATA_ROOT / "raw" / "corpus"
OUTPUT_DIR = DATA_ROOT / "processed" / "corpus"
```

Cargamos todos los archivos ".md" dentro de la carpeta de entrada.

Tras leerlos, aplicamos un par de transformaciones a los documentos para dejarlos en un estado mejor.

```python
for doc_path in sorted(INPUT_DIR.rglob("*.md")):
    relative = doc_path.relative_to(INPUT_DIR)

    species = relative.parts[0]
    file_name = relative.parts[1]

    with open(doc_path, 'r', encoding='utf-8') as f:
        text = f.read()
    
    text = text.replace("Caraterísticas:", "## Características:\n\n")
    text = re.sub(r"\n{3,}", "\n\n", text) # Reemplaza 3 o más espacios por 2 espacios.
    text = re.sub(r"(?<!\n)\n(?!\n)", " ", text) # Limpia roturas de línea en mitad de frases.
    text = re.sub(r'\(\s*(.*?)\s*\)', r'(\1)', text) # Cambia texto con formato "( texto )" a "(texto)".
    
    new_path = OUTPUT_DIR / species / file_name
    os.makedirs(new_path.parent, exist_ok=True)
    with open(new_path, 'w', encoding='utf-8') as f:
        f.write(text)
```
